### Standardize Column Names

The dataset column is in a snake case in most time we can get errorin our code from incosistent column name. its adviseable we conver it to lowercase using:

- .str.strip() → removes extra spaces

- .str.lower() → converts everything to lowercase

- .str.replace(" ", "_") → replaces spaces with underscores

After we run the code the columns are:

- consistent
- easier to type
- production-ready

This avoids errors like:
- df["Operating Airline"]   ❌
- df["operating_airline"]   ✅

```python
df.columns=df.columns.str.strip().str.lower().str.replace(" ","_")
df.columns
```

#### Fix Date Column (CRITICAL & Very Important)

converting the date data type from object(str) to datetime data type using:

- .loc[row_indexer, column_indexer]
- : means "all rows"
- "date" is the column

This conversion helps for :

- time filtering
- resampling
- rolling analysis

Without this, time series analysis is impossible.

```python
df.loc[:, "date"] = pd.to_datetime(df["date"])
df["date"].dtype
```

### checking for missing value

Now we want to understand the data and the aim of this is to profile the data

**Ask:**

- Which columns have missing data?

- How much data is missing?

```python
df.isnull().sum()
```

we can now verify that our dataset has 3 columns with missing values, which are listed below:

- operating_airline_iata_code    75
- published_airline_iata_code    75
- aircraft_manufacturer          11

instead of dropping the column with a missing values we are going to fill it with "na or unknown" and the reason for it:

- it preserve data
- it avoid bias
- it keep dataset size intact 

```python
df.loc[:, "operating_airline_iata_code"] = df["operating_airline_iata_code"].fillna("unknown")
df.loc[:, "published_airline_iata_code"] = df["published_airline_iata_code"].fillna("unknown")
df.loc[:, "aircraft_manufacturer"] = df["aircraft_manufacturer"].fillna("unknown")
```

```python
# we have 11 duplicates on our data Keeping them would double-count data

df = df.drop_duplicates()
```

### Saving A Clean Dataset 

After your data cleaning processess its always a good norm to save the transfomative dataset to your desireable location as its can be helpful when you need to do an explorative data analysis or anything you need to do with the data.

To save your clean dataset in pandas you to:

- import os (operating system)

- Create the datasets directory if it doesn't exist

- save it to the directory you created

- or you can save it from the path where you read(import) the file from the beginning 


```python
import os

# Create the datasets directory if it doesn't exist
if not os.path.exists("datasets"):
    os.makedirs("datasets")
    
# Save our clean data
df.to_csv("datasets/cleaned_air_traffic_2026.csv", index=False)
```
